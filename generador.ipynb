```python
def build_string_numbered(v, q):
    result = []
    for i in range(q):
        result.append(f'{v}{i}')
    return result

def build_triple(o,r,d):
    return "{0},{1},{2}".format(o,r,d)


def append2Hechos(origen,relacion,destino):
    resultado = []
    for i in destino:
        resultado.append(build_triple(origen,relacion,i))
    return resultado

```
